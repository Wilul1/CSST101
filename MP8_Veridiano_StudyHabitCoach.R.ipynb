```r
# Title: Travel Recommender Expert System
destination <- readline(prompt="Do you prefer mountains or beach? ")

if (destination == "mountains") {
  weather <- readline(prompt="Do you like cold weather? (yes/no): ")
  if (weather == "yes") {
    print("Suggested destination: Baguio City.")
  } else {
    print("Suggested destination: Tagaytay.")
  }
} else if (destination == "beach") {
  budget <- as.numeric(readline(prompt="What is your budget in PHP? "))
  if (budget > 5000) {
    print("Suggested destination: Boracay.")
  } else {
    print("Suggested destination: Laiya, Batangas.")
  }
} else {
  print("Please select either mountains or beach.")
}
```

```r
study <- as.numeric(readline("How many hours do you study per day? "))
focus <- readline("Do you get distracted easily? (yes/no): ")
sleep <- as.numeric(readline("How many hours do you sleep? "))

if (study < 2) {
  print("You need to study more.\n")
} else if (study <= 4 && focus == "yes") {
  print("Good effort, but try improving focus.\n")
} else if (study > 4 && sleep < 6) {
  print("You study hard, but get more sleep.\n")
} else if (study > 4 && focus == "no") {
  print("Great discipline! Keep it up.\n")
} else {
  print("Balance study time and rest for best results.\n")
}
```
