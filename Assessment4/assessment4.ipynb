```python
exceptions = ["penguin", "ostrich", "kiwi"]

print("Welcome to the Bird Flying Assumption System!")
animal = input("Enter the animal name: ").lower()

print(f"\nStep 1: Assume {animal} is a bird and can fly.")
if animal in exceptions:
    print(f"Step 2: Check exceptions → {animal} is an exception.")
    print(f"Step 3: Revise conclusion → {animal} cannot fly.")
else:
    print(f"Step 2: No exceptions found.")
    print(f"Step 3: Conclusion → {animal} can fly.")
```
